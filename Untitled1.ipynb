```python
import string
import collections
from collections import defaultdict

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
import re

from wordcloud import WordCloud
from textwrap import wrap
from textblob import TextBlob

#nltk.download('stopwords')
from nltk.corpus import stopwords

OPWORDS = set(stopwords.words('english'))
import pandas as pd 
data1 = pd.read_csv('Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv')
data1.shape
```

```python
data1['reviews.title'] = data1['reviews.title'].apply(lambda x: str(x).lower())
data1['reviews.title'] = data1['reviews.title'].apply(lambda x: re.sub('\w*\d\w*','', x))
#data_df['reviews.title'] = data_df['reviews.title'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))

data1['reviews.text'] = data1['reviews.text'].apply(lambda x: str(x).lower())
data1['reviews.text'] = data1['reviews.text'].apply(lambda x: re.sub('\w*\d\w*','', x))
#data_df['reviews.text'] = data_df['reviews.text'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))
```

```python
STOPWORDS = set(stopwords.words('english'))
data1['reviews.text'] = data1['reviews.text'].apply(lambda x: ' '.join([w for w in x.split(' ') if (w not in STOPWORDS)]))
```
